# Lung Cancer Classification project

This notebook looks into using various Python-based machine learning and data science libraries in an attempt to build a machine learning model capabel of predicting whether or not someone's risk with lung cancer based on their medical attributes.


```python
# Importing tools

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

# Lung Cancer Classification Project

Getting the data ready


```python
# 1. Importing CSV file
lung_cancer = pd.read_csv("cancer patient data sets.csv")
lung_cancer
```

```python
lung_cancer.shape
```

```python
lung_cancer.head
```

```python
# Create a dictionary mapping to fit DF.
level_mapping = {'Low': 1, 'Medium': 2, 'High': 3}

# Apply mapping to the "Level" column
lung_cancer['Level'] = lung_cancer['Level'].map(level_mapping)

# Save the modified DataFrame back to a CSV file
lung_cancer.to_csv('modified_file.csv', index=False)
```

```python
# Extract the numeric part of the Patient Id column and convert it to float
lung_cancer['Patient Id'] = lung_cancer['Patient Id'].str.replace('P', '').astype(float)
```

```python
# Checking data after changes
lung_cancer
```

```python
# Create X (features matrix)
X = lung_cancer.drop("Level", axis=1)

# Create y (labels)
y = lung_cancer["Level"]
```

```python
# Checking to see if it label correctly
X
```

```python
# Checking to see if it is label correctly
y
```

```python
# 2. Choose the righ tmodel and hyperparameters
from sklearn.ensemble import RandomForestClassifier 
clf = RandomForestClassifier(n_estimators=100)

# I'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train, y_train)
```

```python
# Checking training data sets
X_train
```

```python
# Checking training data sets
y_train
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
# Evaluate the model on the training data and the test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

# Checking Classification report, Confusion matrix, and Accuracy Score

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

# Model look too good, let's play around with n_estimators

```python
# Trying to Improving the model
# Try different amount of n_estimators

np.random.seed(42)
for i in range (10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print("")
    
```

```python
# Trying to Improving the model
# Try different amount of n_estimators again***

np.random.seed(42)
for i in range (20, 200, 20):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print("")
    
```

```python
# Saving model
import pickle

pickle.dump(clf, open("random_forest_model_1.pk1", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pk1", "rb"))
loaded_model.score(X_test, y_test)
```
